###dataset

####import all libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

####Read data

```python
df=pd.read_csv('/content/mtcars.csv')
```

```python
df
```

#####Separete the input and output variables

```python
# input -> disp , hp ,wt
x=df[['disp','hp','wt']]
#output ->mpg
y=df[['mpg']]
```

```python
x
```

##### Build the model

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(x,y)
```

```python
#Accuracy
regressor.score(x,y)
```

```python
Disp=221
hp=102
wt=2.91
```

```python
regressor.predict([[Disp,hp,wt]])
```

```python
plt.figure(figsize=(16,9))
plt.subplot(221)
plt.xlabel('disp'); plt.ylabel('mpg')
plt.grid(); plt.title('disp vs. mpg')
plt.scatter(df['disp'],y , color='o')
plt.subplot(222)

plt.xlabel('hp'); plt.ylabel('mpg')
plt.grid();plt.title('hp vs. mpg')
plt.scatter(df['hp'],y, color='b')
plt.subplot(223)

plt.xlabel('wt'); plt.ylabel('mpg')
plt.grid();plt.title('wt vs. mpg')
plt.scatter(df['wt'],y, color='g')
plt.subplot(224)
```
