```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x =np.arange(-10,10,0.25)
y =x*4.23+36.07
plt.plot(x,y)
```

```python
def sigmoid(z):
  val = 1/(1+np.exp(-z))
  return val
```

```python
plt.plot(x,sigmoid(x))
```
