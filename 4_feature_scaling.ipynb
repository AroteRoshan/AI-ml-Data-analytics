#### Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('/content/Social_Media_Ads.csv')
df
```

```python
df.describe()
```

```python
x=df[['Age','EstimatedSalary']]
x
```

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(df['Age'],df['EstimatedSalary'],c=df['Purchased'])
```

```python
import seaborn as sns
sns.jointplot(x='Age',y='EstimatedSalary',data=df,hue='Purchased')
```

```python
from sklearn.preprocessing import MinMaxScaler
#Import the class
```

```python
scaler=MinMaxScaler()
# Create the object
```

```python
x_scaled=scaler.fit_transform(x)
x_scaled
#Apply the method
```

```python
x_scaled=pd.DataFrame(x_scaled,columns=x.columns)
x_scaled
```

```python
x_scaled.describe()
```

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(df['Age'],x_scaled['EstimatedSalary'],c=df['Purchased'])
```

#### Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
x_scaled=pd.DataFrame(x_scaled,columns=x.columns)
x_scaled.describe()
```

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(df['Age'],x_scaled['EstimatedSalary'],c=df['Purchased'])
```
